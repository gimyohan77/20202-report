```python
p = ["상승", "급등", "폭등", "매수",
    "떡상", "가자", "호재", "반등", "목표", "오를" ,"추매", "빨간", "오른", "기대"]
n = ["하락", "급락", "폭락", "매도",
    "떨어", "팔아", "떡락", "악재", "팔고", "조정", "조심", "파란", "내린", "손절"]


def cntpos(text):
    global po
    po = 0
    for i in range(0, 14, 1):
        if text.find(p[i]) >= 0:
            po = po+1
    if po > 0:
        return 1
    if po == 0:
        return 0

def cntneg(text):
    global ne
    ne = 0
    for i in range(0, 14, 1):
        if text.find(n[i]) >= 0:
            ne = ne+1
    if ne > 0:
        return 1
    if ne == 0:
        return 0
```

```python
from konlpy.tag import Twitter
from collections import Counter

def get_tags(text, ntags = 50):
    spliter = Twitter()
    nouns = spliter.morphs(text)
    count = Counter(nouns)
    return_list = []
    for n, c in count.most_common(ntags):
        temp = {'tag':n, 'count':c}
        return_list.append(temp)
    
    return return_list
    
```
